Now for the second part of the analysis the average of the cumulative returns of each closing day should be investigated.First, a new dataframe with the Cumulative returns should be created.

```python
df = pd.DataFrame({'Date': sp.Date,
                   'SP': SP_Cumulative,
                   'Facebook': FB_Cumulative,
                   'Google': Google_Cumulative,
                   'Apple': Apple_Cumulative,
                   'Disney': Disney_Cumulative,
                   'Microsoft': Microsoft_Cumulative})
```

Then the sum of each date should be included in the new dataframe

```python
df['Sum_by_day'] = df['Apple'] + df['Disney'] + df['Facebook'] + df['Google'] + df['Microsoft'] + df['SP']
```

The Total column will include the sums of the cumulatives till each day.

```python
df['Total'] = np.cumsum(df['Sum_by_day'])
```

Now the final column will contain the average of the total of the cumulatives divided by the number of the working days that has passed since the beginning of the year till that day.

```python
n = 0;
df['Average'] = df['Sum_by_day'][0];
for s in range(0,len(df)):
    n = n + 1
    df['Average'][s] = df['Total'][s]/n
```

### Regression

We use regression when we want to examine the relationship between one variable (the dependent variable) and one or more other variables (the independent variables).
In particular we want to find out whether the dependent variable really depends on the independent variables.
In the simplest case, we have just two variables (the dependent and one independent), which we handle with simple linear regression analysis.
Linear regression assumes that the relationship between the dependent variable y and the independent variable x is of the form:
y=α+βx+ϵ
We have a set of yi and xi from our observations, we want to estimate α and β.

### R-suared
What is R squared?
In statistics, the coefficient of determination, denoted R^2 or r^2 and pronounced R squared, is a number that indicates how well data fit a statistical model – sometimes simply a line or a curve. An R^2 of 1 indicates that the regression line perfectly fits the data, while an R^2 of 0 indicates that the line does not fit the data at all. This latter can be because the data is more non-linear than the curve allows, or because it is random.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.graphics as sgr
from matplotlib.collections import LineCollection
from statsmodels.stats.outliers_influence import summary_table
```

First the regression analysis between the S&P Index and Facebook returns

```python
rdf = pd.DataFrame({'Date': sp.Date,
                   'SP': sp.Close,
                   'Facebook': facebook.Close,
                   'Google': google.Close,
                   'Apple': apple.Close,
                   'Disney': disney.Close,
                   'Microsoft': microsoft.Close})
```

### S&P - Facebook

```python
modspfb = smf.ols("SP ~ Facebook", data=rdf)
spfb_res = modspfb.fit()
spfb_res.params
```

```python
spfb_res.summary()
```

In the summary of the first model we can observe that R-squared is 0.008 which is very low the variability of the data is explained only in 0.8% correctly.Furthermore,if we look at the p-value of the coeficient we see that the 0.168 is above 0.05 so the variable of the Facebook Close returns is statistically insignificant to the values of the S&P index. 

The diagram which follows, will contain a simple scatter plot of the observation between the two variables.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.scatter(rdf['SP'],rdf['Facebook'])
ax.set_xlabel('SP')
ax.set_ylabel('Facebook')
```

It is also useful to plot the observed data along with the fitted data, that is, the data on the regression line that correspond to the x values.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = sgr.regressionplots.plot_fit(spfb_res, 1, ax = ax)
```

```python
spv = rdf.SP.values
fbv = rdf.Facebook
fitted = spfb_res.fittedvalues.values
```

What OLS really tries to do is to minimise the squares of the distances of the residuals from the regression line.

```python
fitted_coords = zip(spv, fitted)
data_coords = zip(spv, fbv)
segment_coords = list(zip(fitted_coords, data_coords))
line_segments = LineCollection(segment_coords,
                               colors=('blue'),
                               linewidths=(1),
                               linestyles='solid')

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.add_collection(line_segments)

ax.scatter(spv, fbv)
fig = sgr.regressionplots.abline_plot(model_results=spfb_res, ax=ax, color='r')
ax.set_xlabel('SP')
ax.set_ylabel('Facebook')
```

The results are not good looking regarding the squares of distances of the residuals from the regression line.There is a huge difference between the line and each sum.

Another plot that will give us a clearer view of the regression model is the one containing the fiited values versus the residuals.Generally,The residuals should not show a pattern when plot against the fitted values.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.scatter(fitted, spfb_res.resid)
ax.set_xlabel('fitted')
ax.set_ylabel('residuals')
```

Here in this plot the that no pattern is shown so the results are acceptable since that is the purpose achieved by a regression model.

In OLS, we expect residuals to follow, roughly, the normal distribution with a mean of 0 and some constant variance.
To get an indication of whether residuals come from a normal distribution, we can check for a straight line on a Q-Q plot.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = sm.qqplot(spfb_res.resid, ax=ax)
```

The Q-Q plot results are not satisfactory at all,because it is clear that the residuals do not descend from a normal distribution.
